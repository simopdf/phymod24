```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ce code permet d'afficher les courbes représentant la digue

H = 10
xa = -5
xe = +5


# On définit les fonctions côté eau et côté air, ici les courbes concaves
def fair(y):
    return -(y - H) * xa / H

def feau(y):
    return -(y - H) * xe / H

# On crée un tableau de y
y = np.linspace(0, H, 1000)

# On trace
plt.plot(fair(y), y, '-b') 
plt.plot(feau(y), y, '-b')  ## Décommenter après avoir codé la fonction feau(x)
 
plt.plot([xe, xa], [0, 0], '-b') # Bas de la digue
plt.xlabel('x (m)') 
plt.ylabel('y (m)')
plt.text(xe, 8, 'eau')
plt.text(xa, 8, 'air')
plt.axis('equal') # Permet d'avoir un plot isométrique
plt.axhline(0, color = "black")
plt.axvline(0, color = "black")
plt.xlim(-8, 8)
plt.show()
```

```python
def P(y): #pression exercée par l'eau
    p = 1000*9.81*y
    return p


# -- integrale digue pentes droites

H = 10
xa = -5
xe = +5

a = 0
b = H
n = 1000
dy = (b-a)/n
vol=0
for i in range(n):
    vol = vol - fair(a + i*dy)*dy + feau(a + i*dy)*dy
print("Le volume est ",vol)

Poids = vol*9.81*3000 #volume * g * masse_volumique
print("La force de gravité exercée par la digue est ", Poids)


# -- forces de pression coté eau (cas xa=-5,xe=5)
ya = 0
yb = H
dy=(yb-ya)/n

Fpx=0
Fpy=0
### Integrale selon x
for i in range(n):
    Fpx= Fpx + P(ya+dy*i)*dy
print("La composante selon u_x est Fpx = ",Fpx, "N")

#intégrale 2 -- Fpy
'''La dérivée de la fonction feau2 est dfeau2(y)'''

for i in range(n):
    Fpy=Fpy+P(ya+i*dy)*(-xe/H)*dy
print("La composante selon u_y est Fpy = ",Fpy, "N")






##
N1 = Poids - Fpy
T1 = Fpx

print("N1=",N1,"T1=",T1, "T/N=",T1/N1)
if (T1/N1<0.5):
    print("La digue est stable")
else:
    print("La digue n'est pas stable")

    
    


```

$$
\left\{\begin{matrix}
\vec n = \sin(\theta).\vec{u_y} - \cos(\theta).\vec{u_x} & \\
\vec{F} = \iint_S \vec{\mathrm{d}F} =\int_0^H P(y) \mathrm{d}l . \vec{n} & \\
\mathrm{d}l = \frac{\mathrm{d}y}{\cos(\theta)} & \\
\tan(\theta) = f'(y) & \\
\end{matrix}\right. $$

$$
\vec{F} =\int_0^H P(y) . \frac{\mathrm{d}y}{\cos(\theta)} . [\sin(\theta).\vec{u_y} - \cos(\theta).\vec{u_x} ]
$$

$$
\vec{F} =\int_0^H P(y) . \mathrm{d}y . [\tan(\theta).\vec{u_y} - \vec{u_x} ]
$$

$$
\left\{\begin{matrix}
    \vec{F}_x = - \int_0^H P(y) . \mathrm{d}y . \vec{u_x} & \\
    \vec{F}_y = \int_0^H P(y) . \mathrm{d}y . \tan(\theta).\vec{u_y} & \\
\end{matrix}\right.
 \Rightarrow
\left\{\begin{matrix}
    \vec{F}_x = - \int_0^H P(y) . \mathrm{d}y . \vec{u_x} & \\
    \vec{F}_y = \int_0^H P(y) . \mathrm{d}y . f'(y).\vec{u_y} & \\
\end{matrix}\right.
$$

```python
# Ce code permet d'afficher les courbes représentant la digue

H = 10
xe = 5
xa = 0

# On définit les fonctions côté eau et côté air, ici les courbes concaves
a = 3
def feau(y):
    return a * (y / H) ** 2 - (a + xe) * (y / H) + xe
def fair(y):
    return (H - y) * xa / H

# On crée un tableau de y
y = np.linspace(0, H, 1000)

# On trace
plt.plot(feau(y), y, '-b')
plt.plot(fair(y), y, '-b')
plt.plot([xe, xa], [0, 0], '-b')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.text(3, 8, 'eau')
plt.text(-3, 8, 'air')
plt.axis('equal') # Permet d'avoir un plot isométrique
plt.axhline(0, color = "black")
#plt.axvline(0, color = "black")
plt.xlim(-8, 8)
plt.show()
```

```python
H = 10
n = 1000
xa = 0
xe = 5
#fonction pression, fonction digue et derivée
def P(y,H=10): #pression exercée par l'eau
    p = 1000*9.81*(H-y)
    return p
def feau2(y,xe,H=10,a=3):
    fy = a*(y/H)**2-(a+xe)*(y/H)+xe
    return fy

def dfeau2(y,xe,H=10,a=3):
    dfy =2*a*y/(H**2)-(a+xe)/H
    return dfy
```

```python
### Cas xe = 5
x_e=5
# volume de beton
start = 0
end = H
dy=(end-start)/n
volc = 0
n=1000


for i in range(n):
    volc = volc + feau2(start + i*dy,x_e)*dy
print("Le volume est ", volc)

#forces de pression
Fpxc = 0
Fpyc = 0
##intégrale 1 -- Fpx
for i in range(n):
    Fpxc= Fpxc + P(start+dy*i)*dy
print("La composante selon u_x est Fpx = ",Fpxc, "N")

#intégrale 2 -- Fpy
'''La dérivée de la fonction feau2 est dfeau2(y)'''

for i in range(n):
    Fpyc=Fpyc+P(start+i*dy)*dfeau2(start+i*dy,x_e)*dy
print("La composante selon u_y est Fpy = ",Fpyc, "N")

#bilan des forces
poids2 = -volc*9.81*3000 #volume * g * masse_volumique

N2 = Fpyc + poids2 #Selon y on a le poids et la composante Fpy des forces de pression
T2= -Fpxc #Selon x on a la composante Fpx des forces de pression et la force de frottement (opposée à Fpy). On suppose équilibre donc T+Fpx=0


print("N = " , N2)
print("T =", T2)
print("T/N = ", np.abs(T2/N2))

equilibre = (np.abs(T2/N2)<=0.5)
if equilibre:
    print("La digue est à l'équilibre")
else:
    print("La digue n'est pas stable")
```

```python
####Fonction (T/N) ###
def stable(x_e,H=10,start=0,end=H,n=1000):
    volc = 0
    dy=(end-start)/n
    for i in range(n):
        volc = volc + feau2(start + i*dy,x_e)*dy
    #print("Le volume est ", volc)

    #forces de pression
    Fpxc = 0
    Fpyc = 0
    ##intégrale 1 -- Fpx
    for i in range(n):
        Fpxc= Fpxc + P(start+dy*i)*dy
    #print("La composante selon u_x est Fpx = ",Fpxc, "N")

    #intégrale 2 -- Fpy
    '''La dérivée de la fonction feau2 est dfeau2(y)'''

    for i in range(n):
        Fpyc=Fpyc+P(start+i*dy)*dfeau2(start+i*dy,x_e)*dy
    #print("La composante selon u_y est Fpy = ",Fpyc, "N")

    #bilan des forces
    poids2 = -volc*9.81*3000 #volume * g * masse_volumique

    N2 = Fpyc + poids2 #Selon y on a le poids et la composante Fpy des forces de pression
    T2= -Fpxc #Selon x on a la composante Fpx des forces de pression et la force de frottement (opposée à Fpy). On suppose équilibre donc T+Fpx=0
    mu = np.abs(T2/N2)
    return volc,mu
```

```python
# Stabilité pour xe in [5,6]
xe_tab = np.linspace(5,6,300)

for x_e in xe_tab:
    V,M=stable(x_e)
    plt.scatter(V,M, marker =".", s=1,c="r")
plt.xlabel("Volume")
plt.ylabel("$T/N$")
plt.plot(np.linspace(20,25,xe_tab.size),np.ones_like(xe_tab)*0.5, label="limite de stabilité. $T/N=0.5$")
plt.legend()
```
